# Mean Normalization

```python
def mean_normalization(df):
    cont_columns = df.select_dtypes(include=[np.float64, np.float32, np.int64, np.int32]).columns
    df_cont = (df[cont_columns] - df[cont_columns].mean()) / (df[cont_columns].max() - df[cont_columns].min())
    cat_columns = df.select_dtypes(include=['category', 'object']).columns
    data = pd.concat([df_cont, df[cat_columns]], axis=1)
    return data
```

```python
new_df = mean_normalization(df)
```
